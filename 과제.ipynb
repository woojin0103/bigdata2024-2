```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager
plt.rcParams.update({'figure.dpi':'150'})
plt.rcParams.update({'figure.figsize':[6,4]})
plt.rcParams.update({'font.size' :'15'})
plt.rcParams.update({'font.family' :'Malgun Gothic'})
plt.rcParams['axes.unicode_minus'] = False
```

```python
고양시등록 = pd.read_csv('경기도 고양시_등록장애인현황_20241104.csv',encoding = 'EUC-KR')
고양시등록 = 고양시등록.drop('읍면동명',axis=1)
고양시등록 = 고양시등록.drop(고양시등록.index[-1])
고양시등록['등록장애인수(합계)'] = 고양시등록['등록장애인수(합계)'].replace({',': ''}, regex=True).astype(int)
고양시등록 = 고양시등록[고양시등록['장애유형'] != '소계']
고양시등록
```

```python
plt.figure(figsize=(10,6))
plt.bar(고양시등록['장애유형'], 고양시등록['등록장애인수(합계)'], color='skyblue')

plt.title('고양시 장애유형별 등록장애인수(합계)', fontsize=14)
plt.xlabel('장애유형', fontsize=12)
plt.ylabel('등록장애인수(합계)', fontsize=12)

plt.xticks(rotation=45, ha="right") 
plt.tight_layout()
plt.show()
```

```python
성남시등록 = pd.read_csv('경기도 성남시_장애인등록_현황_20240614.csv',encoding='EUC-KR')
성남시등록 = 성남시등록.drop(['읍면동명','데이터기준일자'],axis=1)
성남시등록
```

```python
plt.figure(figsize=(10,6))
plt.bar(성남시등록['장애유형'], 성남시등록['등록장애인수(합계)'], color='skyblue')

plt.title('성남시 장애유형별 등록장애인수(합계)', fontsize=14)
plt.xlabel('장애유형', fontsize=12)
plt.ylabel('등록장애인수(합계)', fontsize=12)

plt.xticks(rotation=45, ha="right")  
plt.tight_layout()
plt.show()
```

```python
고양성남등록 = pd.concat([고양시등록,성남시등록])
고양성남등록 = 고양성남등록.reset_index(drop=True)
고양성남등록
```

```python
plt.figure(figsize=(10, 6))

plt.barh(고양시등록["장애유형"], 고양시등록["등록장애인수(합계)"], color="blue", label="고양시")
plt.barh(성남시등록["장애유형"], -성남시등록["등록장애인수(합계)"], color="green", label="성남시")

plt.title("고양시와 성남시 장애유형별 등록 장애인 수 비교", fontsize=14)
plt.xlabel("등록 장애인 수", fontsize=12)
plt.ylabel("장애유형", fontsize=12)
plt.legend()

plt.show()
```

```python
장애인편의시설 = pd.read_csv('복지시설시설유형별시설수.csv',encoding='EUC-KR')
장애인편의시설
```

```python
고양성남편의 = 장애인편의시설[(장애인편의시설["시군구명"] == "고양시") | (장애인편의시설["시군구명"] == "성남시")]
고양성남편의
```

```python
고양시편의 = 장애인편의시설[장애인편의시설["시군구명"] == "고양시"]
고양시편의 = 고양시편의.reset_index(drop=True)
고양시편의
```

```python
성남시편의 = 장애인편의시설[장애인편의시설["시군구명"] == "성남시"]
성남시편의 = 성남시편의.reset_index(drop=True)
성남시편의
```

```python
고양시편의['복지시설수'].sum()
```

```python
성남시편의['복지시설수'].sum()
```

```python
goyang_df = 장애인편의시설[장애인편의시설["시군구명"] == "고양시"]
seongnam_df = 장애인편의시설[장애인편의시설["시군구명"] == "성남시"]

goyang_total = goyang_df["복지시설수"].sum()
seongnam_total = seongnam_df["복지시설수"].sum()

total = goyang_total + seongnam_total

labels = ['고양시', '성남시']
sizes = [goyang_total, seongnam_total]
colors = ['blue', 'green']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=85)
plt.title("고양시와 성남시 전체 복지시설 수 비교", fontsize=14)

plt.tight_layout()
plt.show()
```

```python
goyang_ratio = 고양시편의["복지시설수"].sum() / 고양시등록["등록장애인수(합계)"].sum()
seongnam_ratio = 성남시편의["복지시설수"].sum() / 성남시등록["등록장애인수(합계)"].sum()

labels = ['고양시', '성남시']
ratios = [goyang_ratio.mean(), seongnam_ratio.mean()]
colors = ['blue', 'green']

plt.pie(ratios, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)

plt.title("고양시와 성남시 복지시설 수/등록 장애인 비율 비교", fontsize=14)

plt.show()
```
